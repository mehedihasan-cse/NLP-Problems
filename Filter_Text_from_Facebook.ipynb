```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ESUIT _ Comments Exporter for Facebook‚Ñ¢ (42).csv')
```

```python
df = df.drop(columns = ['Id', 'UserId', 'Avatar', 'Author', 'Url', 'ReactionsCount',
       'Depth', 'SubCommentsCount', 'CommentAt'])
df
```

```python
df.dropna()
```

```python
import re

def has_english(row):
    # Check if the 'Content' column contains any English letters
    content = row['Content']
    if isinstance(content, str):  # Check if content is a string
        return bool(re.search(r'[a-zA-Z]', content))
    else:
        return False  # Return False for non-string values

# Filter the DataFrame to keep only rows without English letters
df = df[~df.apply(has_english, axis=1)]
df
```

```python
import string
punct=string.punctuation
```

```python
def remove_punct(x):
    if isinstance(x, str):  # Check if x is a string
        return x.translate(str.maketrans("", "", punct))
    else:
        return x  # Return the original value if it's not a string

df['Content'] = df['Content'].apply(lambda x: remove_punct(x))
```
